```python
sub = pd.read_csv(r"C:\Users\imane\OneDrive\Desktop\home_exchange\HomeExchange\data\subscriptions_clean.csv")
sub.head()
```

```python
sub = sub.drop(columns=["Unnamed: 0"])
```

```python
sub = sub.rename(columns={"subscription_date_min":"second_subscription","subscription_date_max":"last_subscription",
                         "subscription_date_count":"nb_year","renew_last":"has_renewed",
                         "first_subscription_date_first": "first_suscription_date","first_subscription_last":"is_new_user",
                         "referral_first":"referral","promotion_first":"promotion","payment3x_first":"payment3x",
                         "payment2_last":"payement2","payment3_last":"payment3"})
```

```python
sub["second_subscription"] = pd.to_datetime(sub["second_subscription"])
sub["last_subscription"] = pd.to_datetime(sub["last_subscription"])
sub["first_suscription_date"] = pd.to_datetime(sub["first_suscription_date"])
```

```python
sub["abscence_year"] = sub["second_subscription"] - sub["first_suscription_date"]
sub['abscence_year'] = pd.to_numeric(sub['abscence_year'].dt.days, downcast='integer')
```

```python
exc = pd.read_csv(r"C:\Users\imane\OneDrive\Desktop\home_exchange\HomeExchange\data\exchanges_clean_final.csv")
exc.head()
```

```python
sub.info()
```

```python
exc.info()
```

```python
exc = exc.drop(columns=["Unnamed: 0"])
exc["created_at"] = pd.to_datetime(exc["created_at"])
exc["start_on"] = pd.to_datetime(exc["start_on"])
exc["end_on"] = pd.to_datetime(exc["end_on"])
```

```python
exc["exchange_type"] = exc["exchange_type"].astype("string")
exc["home_type"] = exc["home_type"].astype("string")
exc["residence_type"] = exc["residence_type"].astype("string")
exc["country"] = exc["country"].astype("string")
```

```python
guests = exc.groupby("guest_user_id").agg({"host_user_id":"count", "guest_countguest_count": "mean","night_count":"mean",
                                          "exchange_type":pd.Series.mode,"home_type":pd.Series.mode,"residence_type":pd.Series.mode,
                                          "capacity":"mean","diff_capacity":"mean","is_finalized":"sum","diff_book_stay":"mean"}).reset_index()
```

```python
guests = guests.rename(columns={"host_user_id":"nb_of_exchanges"})
```

```python
guests = guests.round(0)
```

```python
hosts = exc.groupby("host_user_id").agg({"guest_user_id":"count", "guest_countguest_count": "mean","night_count":"mean",
                                          "exchange_type":pd.Series.mode,"home_type":pd.Series.mode,"residence_type":pd.Series.mode,
                                          "capacity":"mean","diff_capacity":"mean","is_finalized":"sum","diff_book_stay":"mean"}).reset_index()
```

```python
hosts = hosts.rename(columns={"guest_user_id":"nb_of_exchanges"})
```

```python
hosts = hosts.round(0)
```

```python
guests = guests.rename(columns={"guest_user_id":"user_id","nb_of_exchanges":"exchanges_guest","night_count":"nights_guest",
                             "exchange_type":"exchange_type_guest","home_type":"home_guest","residence_type":"residence_guest",
                               "capacity":"capacity_guest","diff_capacity":"diff_capacity_guest","is_finalized":"finalized_guest",
                               "diff_book_stay":"book_diff_guest","guest_countguest_count":"nb_guests_guest"})
```

```python
hosts = hosts.rename(columns={"host_user_id":"user_id","nb_of_exchanges":"exchanges_host","night_count":"nights_host",
                             "exchange_type":"exchange_type_host","home_type":"home_host","residence_type":"residence_host",
                               "capacity":"capacity_host","diff_capacity":"diff_capacity_host","is_finalized":"finalized_host",
                               "diff_book_stay":"book_diff_host","guest_countguest_count":"nb_guests_host"})
```

```python
sub = sub.rename(columns={"user_id_":"user_id"})
```

```python
sub_guest = sub.merge(guests, on="user_id",how="left")
```

```python
sub_guest.head()
```

```python
sub_guest[["exchanges_guest","nb_guests_guest","nights_guest",
           "capacity_guest","diff_capacity_guest","finalized_guest","book_diff_guest"]] = sub_guest[["exchanges_guest","nb_guests_guest","nights_guest",
           "capacity_guest","diff_capacity_guest","finalized_guest","book_diff_guest"]].fillna(0)
```

```python
sub_guest[["exchange_type_guest","home_guest","residence_guest"]] = sub_guest[["exchange_type_guest",
                                                                               "home_guest","residence_guest"]].fillna("NOT GUEST")
```

```python
df_all = sub_guest.merge(hosts, on="user_id",how="left")
```

```python
df_all[["exchanges_host","nb_guests_host","nights_host",
           "capacity_host","diff_capacity_host","finalized_host","book_diff_host"]] = df_all[["exchanges_host","nb_guests_host","nights_host",
           "capacity_host","diff_capacity_host","finalized_host","book_diff_host"]].fillna(0)
```

```python
df_all[["exchange_type_host","home_host","residence_host"]] = df_all[["exchange_type_host",
                                                                      "home_host","residence_host"]].fillna("NOT HOST")
```

```python
df_all.head()
```

```python
df_all = df_all.drop(columns=["second_subscription",'last_subscription','first_suscription_date'])
```
